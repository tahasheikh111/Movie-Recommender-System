```python
import ast
ast.literal_eval
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l;
```

```python
print(convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'))
movies['genres'].apply(convert)
movies['genres']=movies['genres'].apply(convert)
```

```python
def convert_cast(obj):
    l=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            l.append(i['name'])
            count=count+1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies.head(1)
```

```python
movies['crew'][0]
```

```python
def convert_crew(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break;
    return l
```

```python
movies['crew']=movies['crew'].apply(convert_crew)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tag']=movies['overview']+movies['keywords']+movies['genres']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['id','title','tag']]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def ownstem(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l) # convert list into string again
```

```python
new_df.head()
```

```python
new_df['tag']=new_df['tag'].apply(lambda x:" ".join(x))
new_df['tag'][0]
new_df['tag']=new_df['tag'].apply(ownstem)
new_df['tag']=new_df['tag'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tag']).toarray()
vectors[0]
```

```python
cv.get_feature_names_out()[:100]
```

```python
ps.stem('activities')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
#0 index 1 islya is ka apny sath 1 hi hoga
#index bhi save kerny bcz sorted kerna index chnage hogaye gy 
#thats why
list(enumerate(similarity[0]))
#now apply sorting 
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x: x[1])[1:10]
```

```python
def recommend(movie):
    #fetch index 
    movie_index=new_df[new_df['title']==movie].index[0]
    #now sorted the 1 to 0 similarity index index position pakar y rakhna
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x: x[1])[1:10]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```
